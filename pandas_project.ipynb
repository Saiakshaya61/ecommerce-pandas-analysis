```python
import pandas as pd
df=pd.read_excel("ecommerce data.xlsx")
```

```python
df  #information present in excel
```

```python
df.info()   #Data Understanding
```

```python
df.describe()  #Data Understanding
```

```python
df.isnull().sum()
```

```python
df=df.dropna()
df
```

```python
df["Total_Amount"] = df["Price"] * df["Quantity"]
df
```

**ANALYSIS PART**

CALUCLATING THE TOTAL REVENUE

```python
df['Total_Amount'].sum()
```

CHECKING TOP PRODUCTS

```python
df.groupby("Product")["Quantity"].sum().sort_values(ascending=False)
```

Revenue by city

```python
df.groupby("City")["Total_Amount"].sum()
```

CHECKING TOP CUSTOMERS BASED ON AMOUNT

```python
df.groupby("Customer_ID")["Total_Amount"].sum().sort_values(ascending=False).head()
```

PAYMENT METHOD USAGE

```python
df["Payment_Method"].value_counts()
```

CATEGORY ANALYSIS

```python
df.groupby("Category")["Total_Amount"].sum().sort_values(ascending=False)
```

TIME-BASED ANALYSIS

```python
df["Date"] = pd.to_datetime(df["Date"])
df
```

```python
df.groupby("Date")["Total_Amount"].count()
```

```python
df.groupby(df["Date"].dt.month)["Total_Amount"].sum()
```

```python
df.groupby(df["Date"].dt.month)["Total_Amount"].count()
```

LOCATION ANALYSIS

```python
df.groupby("City")["Total_Amount"].sum().sort_values(ascending=False)
```

```python
df[df["Total_Amount"] > 10000]# High-value orders
```

TOP N ANALYSIS

```python
df.sort_values(by="Total_Amount", ascending=False).head(5)
```
